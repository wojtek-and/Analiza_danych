```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.dropna(thresh=2,inplace=True)
```

```python
df['DOB'] = pd.to_datetime(df['DOB'],format='%m/%d/%y')
df['DateofTermination'] = pd.to_datetime(df['DateofTermination'],format='%m/%d/%y')
df['DateofHire'] = pd.to_datetime(df['DateofHire'],format='%m/%d/%Y')
```

```python
import datetime as dt

def count_seniority(row):

    if pd.isnull(row['DateofTermination']):
        end_date = dt.datetime(2019,9,27)
    else:
        end_date = row['DateofTermination']

    return (end_date - row['DateofHire'])/np.timedelta64(1,'W')
```

```python
df['Seniority'] = df.apply(lambda row: count_seniority(row),axis=1)
```

```python
df[['RecruitmentSource','Seniority']].head()
```

```python
summary = df.groupby('RecruitmentSource')['Seniority'].sum().reset_index()
summary.columns = ['RecruitmentSource', 'TotalSeniority']
summary
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='TotalSeniority', y='RecruitmentSource', data=summary)
```

```python
max_seniority_row = summary.loc[summary['TotalSeniority'].idxmax()]
max_recruitment_source = max_seniority_row['RecruitmentSource']
```

```python
print(f"Najlepszym źródłem pozyskania pracownika, jeśli zależy nam na jak najdłuższym stażu pracowników jest {max_recruitment_source}.")
```

```python
plt.figure(figsize=(4,5))
sns.barplot(x='MaritalDesc',y='EmpSatisfaction',data=df)
```

```python
print("Na podstawie powyższego wykresu zależności stanu cywilnego od zadowolenia z pracy można stwierdzić, że pomiędzy tymi zmiennymi nie występuje wyraźna korelacja. Wszystkie wartości zadowolenia w podziale na stan cywilny są do siebie zbliżone i oscylują w granicach 4.")
```

```python
def count_age(row, reference_date):
    if pd.isnull(row['DateofTermination']) and row['DOB'] < reference_date:
        return (reference_date - row['DOB'])/np.timedelta64(1,'W')
    else:
        return None

df['Age'] = df.apply(lambda row: count_age(row, dt.datetime(2019,9,27)),axis=1)
```

```python
sns.distplot(df['Age'])
```

```python
print("Najwięcej zatrudnionych pracowników jest w wieku około 1750 tygodni, czyli około 34 lat.")
print("Funkcja 'timedelta' nie przyjmuje niestety argumentu 'Y' jako lata, dlatego zdecydowałem się na wyświetlenie wieku w tygodniach.")
```

```python
sns.jointplot(x='Age',y='SpecialProjectsCount',data=df)
```

```python
print("Z powyższego wykresu wynika, że raczej młodsi pracownicy pracują nad większą liczbą specjalnych projektów niż starsi. Większe zgrupowanie obserwacji występuje przy mniejszym wieku, również podanym w tygodniach. Warto także zwrócić uwagę na wykresy słupkowe pokazujące, że w przypadku młodszych pracowników projektów jest więcej. Przy czym równie warte uwagi jest to, że najwięcej obserwacji całej populacji dotyczy sytuacji gdzie pracownik nie realizuje żadnego projektu specjalnego.")
```
